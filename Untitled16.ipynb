```python
# 1. Εγκατάσταση βιβλιοθηκών
!pip install -q pymupdf transformers[sentencepiece] sacremoses

import fitz  # PyMuPDF
from transformers import MarianMTModel, MarianTokenizer
import torch
from google.colab import files

# 2. Προετοιμασία AI Μοντέλου (Αγγλικά -> Ελληνικά)
model_name = "Helsinki-NLP/opus-mt-en-el"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_safe(text):
    if not text.strip(): return ""
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# 3. Ανέβασμα και Διάβασμα PDF
print("Ανέβασε το PDF αρχείο σου:")
uploaded = files.upload()

for filename in uploaded.keys():
    doc = fitz.open(filename)

    # Έλεγχος για κλείδωμα
    if doc.is_encrypted:
        password = input(f"Το αρχείο {filename} είναι κλειδωμένο. Δώσε κωδικό: ")
        if not doc.authenticate(password):
            print("Λάθος κωδικός!")
            continue

    print(f"\nΔιάβασμα αρχείου: {filename}")

    # Παίρνουμε το κείμενο από την πρώτη σελίδα ως δείγμα
    first_page_text = doc[0].get_text()

    # Χωρίζουμε σε προτάσεις (με την τελεία)
    sentences = first_page_text.split('.')[:5] # Παίρνουμε τις πρώτες 5 προτάσεις για ταχύτητα

    print("-" * 30)
    print("ΜΕΤΑΦΡΑΣΗ ΔΕΙΓΜΑΤΟΣ (Πρώτες 5 προτάσεις):")
    for s in sentences:
        if len(s.strip()) > 5:
            greek = translate_safe(s.strip())
            print(f"\nEN: {s.strip()}")
            print(f"GR: {greek}")

    doc.close()
```
