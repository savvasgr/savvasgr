```python
# 1. Εγκατάσταση των απαραίτητων (αfrom transformers import MarianMTModel, MarianTokenizer
import torch
from google.colab import files

# 1. Φόρτωση Μοντέλου και Tokenizer
model_name = "Helsinki-NLP/opus-mt-en-el"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_safe(text):
    if not text.strip(): return ""
    inputs = tokenizer(text, return_tensors="pt", padding=True)
    with torch.no_grad():
        translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# 2. Ανέβασμα Αρχείου
print("Ανέβασε το αγγλικό αρχείο (.txt):")
uploaded = files.upload()

for filename in uploaded.keys():
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    print(f"Μετάφραση του {filename}...")
    greek_lines = []

    # Μετάφραση ανά γραμμή για να μην "μπουκώσει" η μνήμη
    for line in lines:
        if line.strip():
            translated_line = translate_safe(line.strip())
            greek_lines.append(translated_line)
        else:
            greek_lines.append("") # Διατήρηση κενών γραμμών

    # 3. Αποθήκευση και Κατέβασμα
    output_file = "translated_file_GR.txt"
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("\n".join(greek_lines))

    print("Η μετάφραση ολοκληρώθηκε!")
    files.download(output_file)
```
